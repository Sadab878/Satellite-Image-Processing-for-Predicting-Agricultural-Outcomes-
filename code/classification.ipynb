**Training**

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n-cls.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data=r"./Dataset/veg_images", epochs=100, imgsz=256, project="Veg", name="veg_yolov11_cls")

model.save("veg_yolov11_cls.pt")
```

**Validation**

```python
from ultralytics import YOLO

# Load a model
model = YOLO("veg_yolov11_cls.pt")  # pretrained YOLO11n model

# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.top1  # top1 accuracy
metrics.top5  # top5 accuracy
```
