**Contribution of Each Channel**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in color (BGR)
image = cv2.imread(r"C:\Users\SADAB KHAN\OneDrive\Desktop\image data set\3\3.3\cir.3.3.tif")
# image brightness
brightness_factor = 6.0  # you can adjust this value
image = np.clip(image * brightness_factor, 0, 255).astype(np.uint8)
# Split channels
blue, green, red = cv2.split(image)

# Sum of pixel intensities
sum_blue = np.sum(blue)
sum_green = np.sum(green)
sum_red = np.sum(red)

# Total sum
total = sum_blue + sum_green + sum_red

# Calculate contributions
pct_blue = (sum_blue / total) * 100
pct_green = (sum_green / total) * 100
pct_red = (sum_red / total) * 100

# Print results
print(f"Blue Channel Contribution:  {pct_blue:.2f}%")
print(f"Green Channel Contribution: {pct_green:.2f}%")
print(f"Red Channel Contribution:   {pct_red:.2f}%")

# Determine dominant channel
channels = {'Blue': pct_blue, 'Green': pct_green, 'Red': pct_red}
dominant = max(channels, key=channels.get)
print(f"\nDominant Channel: {dominant} ({channels[dominant]:.2f}%)")

# Bar chart
labels = ['Blue', 'Green', 'Red']
values = [pct_blue, pct_green, pct_red]
colors = ['blue', 'green', 'red']

plt.figure(figsize=(6, 5))
plt.bar(labels, values, color=colors, edgecolor='black')
plt.title('Color Channel Contribution (%)')
plt.ylabel('Percentage')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Highest Red Intensities**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread(image_path)

# Extract Red channel
red = image[:, :, 2]

# Flatten red channel to 1D
red_flat = red.flatten()

# Define bins: 5 equal intensity parts
bins = [(0, 50), (51, 100), (101, 150), (151, 200), (201, 255)]

# Count total pixels
total_pixels = red_flat.size

# Count and percentage in each bin
bin_counts = [np.sum((red_flat >= low) & (red_flat <= high)) for (low, high) in bins]
bin_percentages = [(count / total_pixels) * 100 for count in bin_counts]

# Print text results
print("Red Channel Intensity Contribution:")
for i, ((low, high), pct) in enumerate(zip(bins, bin_percentages), 1):
    print(f"  Bin {i}: Intensities {low}-{high} → {pct:.2f}%")

# Plot bar chart
labels = [f"{low}-{high}" for (low, high) in bins]

plt.figure(figsize=(8, 5))
plt.bar(labels, bin_percentages, color='red', edgecolor='black')
plt.title("Red Channel Intensity Distribution")
plt.xlabel("Intensity Range")
plt.ylabel("Percentage (%)")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
